```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://fr.actualitix.com/blog/densite-de-la-population-a-paris.html#:~:text=La%20moyenne%20de%20la%20densit%C3%A9,par%20l'INSEE%20en%202007."
html_text = requests.get(url).text
```

```python
soup_obj = BeautifulSoup(html_text, "html.parser")
tables = soup_obj.select('table')
table = tables[0]
table_html = str(table)
table_df_list = pd.read_html(table_html)
table_df = table_df_list[0]
table_df.reset_index(drop=True, inplace=True)
```

```python
table_df.columns=['arr','density']
```

```python
table_df['arr'] = table_df['arr'].map({'Paris 1er':75001,
                             'Paris 2e':75002,
                             'Paris 3e':75003,
                                 'Paris 4e':75004,
                                 'Paris 5e':75005,
                                 'Paris 6e':75006,
                                 'Paris 7e':75007,
                                 'Paris 8e':75008,
                                 'Paris 9e':75009,
                                 'Paris 10e':75010,
                                 'Paris 11e':75011,
                                 'Paris 12e':75012,
                                 'Paris 13e':75013,
                                 'Paris 14e':75014,
                                 'Paris 15e':75015,
                                 'Paris 16e':75016,
                                  'Paris 17e':75017,
                                 'Paris 18e':75018, 
                                 'Paris 19e':75019,
                               'Paris 20e':75020},
                             na_action=None)
```

```python
table_df['density'] = table_df['density'].str.replace(' ','')
```

```python
table_df=table_df.drop(table_df.index[20])
table_df['density'] = table_df['density'].str.replace(',','.')
table_df = table_df.astype({'arr': 'int'})
table_df = table_df.astype({'density': 'float'})
```
